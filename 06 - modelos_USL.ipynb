```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score

from sklearn.cluster import KMeans, SpectralClustering, AgglomerativeClustering, DBSCAN
```

```python
train = pd.read_csv('trainP.csv')
```

```python
test = pd.read_csv('testP.csv')
```

```python
test.shape, train.shape
```

```python
X = train.drop(['RENDIMIENTO_GLOBAL', 'ID'],axis=1)
```

```python
X
```

```python
Xtest = test.drop(['ID'], axis=1)
```

```python
y = train['RENDIMIENTO_GLOBAL'].values
```

```python
print(np.unique(X))
print(np.unique(y))
```

```python
# K-Means clustering
estimator1 = KMeans(n_clusters=4)
estimator1.fit(X)
```

```python
prediction = estimator1.predict(Xtest)
prediction.shape
```

```python
estimator1.score(X, y)
```

```python
estimator1.cluster_centers_
```

```python
idTest = pd.DataFrame(test["ID"])

Reference = ['bajo', 'medio-bajo', 'medio-alto', 'alto']
y_preditionReference = [Reference[int(i)] for i in prediction]

idTest['RENDIMIENTO_GLOBAL'] = pd.Series(y_preditionReference, name='RENDIMIENTO_GLOBAL')
```

```python
idTest.to_csv('submissionKm.csv', index=False)
```
